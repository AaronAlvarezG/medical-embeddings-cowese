## 2 Función para preprocesamiento y tokenización
El preprocesamiento incluye:
- Convertir a minúsculas
- Eliminar caracteres especiales, números y puntuación
- Tokenizar por palabras
- Filtrar tokens muy cortos (< 2 caracteres)
- No eliminamos stopwords médicas porque palabras como 'no', 'sin', 'con' son semánticamente importantes en contextos clínicos (ej: 'sin fiebre', 'no alérgico').

```python
# conservamos letras (incluyendo acentos y ñ) y espacios
RE_LIMPIAR = re.compile(r'[^a-záéíóúüñàèìòùâêîôûA-ZÁÉÍÓÚÜÑ\s]')

def preprocesar_linea(texto: str) -> list:
    """
    Toma una línea de texto y devuelve una lista de tokens limpios.
    Pasos:
      1. Minúsculas
      2. Quitar caracteres no alfabéticos
      3. Tokenizar por espacios
      4. Filtrar tokens de 1 caracter
    """
    texto = texto.lower()
    texto = RE_LIMPIAR.sub(' ', texto)
    tokens = texto.split()
    tokens = [t for t in tokens if len(t) > 1]
    return tokens
```

## 3 Generador de oraciones (streaming)

Lee línea por línea. 

```python
class CorpusStreaming:
    """
    lee el corpus línea por línea.
    """
    def __init__(self, ruta: str, max_lineas: int = None):
        self.ruta = ruta
        self.max_lineas = max_lineas  # None = usar todo el corpus

    def __iter__(self):
        with open(self.ruta, 'r', encoding='utf-8', errors='ignore') as f:
            for i, linea in enumerate(f):
                if self.max_lineas and i >= self.max_lineas:
                    break
                tokens = preprocesar_linea(linea)
                if len(tokens) >= 3:  # Ignoramos líneas muy cortas
                    yield tokens

# Instanciamos el corpus completo
corpus = CorpusStreaming(CORPUS_PATH)
```

## 4 Callback para monitorear el entrenamiento

```python
class MonitorEpochs(CallbackAny2Vec):
    """imprime el progreso"""
    def __init__(self, nombre_modelo: str):
        self.nombre = nombre_modelo
        self.epoch = 0
        self.inicio = None
        
    def on_epoch_begin(self, model):
        self.inicio = time.time()
        print(f'  [{self.nombre}] Epoch {self.epoch + 1} iniciando...', end=' ')

    def on_epoch_end(self, model):
        elapsed = time.time() - self.inicio
        print(f'OK! | Tiempo: {elapsed:.1f}s')
        self.epoch += 1
```

## 5 Entrenamiento Word2Vec

## Justificación de los Parámetros Fijos

En este trabajo, los parámetros `window`, `min_count` y `sample` se fijaron con base en recomendaciones establecidas en la literatura sobre Word2Vec, con el fin de aislar el efecto de la dimensión del vector (`vector_size`) y del tipo de arquitectura (CBOW y Skip-gram).

El tamaño de ventana (`window = 7`) permite capturar relaciones semánticas relevantes sin introducir ruido excesivo, favoreciendo la representación de conceptos en textos biomédicos [1], [2].

El parámetro `min_count = 5` se seleccionó para equilibrar la cobertura del vocabulario y la eliminación de términos poco informativos, conservando palabras técnicas de baja frecuencia [3], [4].

El subsampling (`sample = 1e-4`) reduce la influencia de palabras altamente frecuentes, evitando que dominen el proceso de entrenamiento y mejorando la calidad semántica de los embeddings [1], [2].

Asimismo, se utilizó negative sampling (`negative = 10`, `hs = 0`) por su eficiencia computacional y buen desempeño en corpus de gran tamaño [1], [5]. El número de épocas (`epochs = 10`) se fijó por su equilibrio entre convergencia y costo computacional [2], [6].



## Referencias

[1] T. Mikolov et al., "Efficient Estimation of Word Representations in Vector Space," ICLR, 2013.

[2] O. Levy, Y. Goldberg, and I. Dagan, "Improving Distributional Similarity with Lessons Learned from Word Embeddings," TACL, 2015.

[3] F. Soares et al., "Medical Word Embeddings for Spanish," BioNLP, 2019.

[6] R. Rehurek and P. Sojka, "Software Framework for Topic Modelling with Large Corpora," LREC, 2010.



```python
# Valores a experimentar
VECTOR_SIZES = [50, 100, 200, 300]
SG_VALUES = [0, 1]  # 0=CBOW, 1=Skip-gram

# Parámetros fijos 
FIXED_PARAMS = dict(
    window=7,
    min_count=5,
    sample=1e-4,
    hs=0,
    negative=10,
    ns_exponent=0.75,
    epochs=10,
    workers=20,
    seed=42,
    compute_loss=False 
)



# FUNCIÓN DE ENTRENAMIENTO
def train_and_save(corpus, vector_size, sg):

    model_name = f"w2v_D{vector_size}_sg{sg}.txt"
    out_path = os.path.join(OUTPUT_DIR, model_name)

    print("=" * 60)
    print(f"Entrenando: D={vector_size} | sg={sg}")
    print(f"Archivo: {model_name}")

    start_time = time.time()
    
    monitor_w2v = MonitorEpochs('Word2Vec')
    
    # Crear modelo
    model = Word2Vec(
        sentences=corpus,
        vector_size=vector_size,
        sg=sg,
        **FIXED_PARAMS,
        callbacks=[monitor_w2v]
    )

    # Guardar solo vectores en formato TXT
    model.wv.save_word2vec_format(
        out_path,
        binary=False
    )

    elapsed = (time.time() - start_time) / 60

    print(f"Finalizado en {elapsed:.2f} minutos")
    print(f"Vocabulario: {len(model.wv)} palabras")

    # Liberar memoria
    del model
    gc.collect()



# LOOP EXPERIMENTAL
def run_experiments(corpus):

    total = len(VECTOR_SIZES) * len(SG_VALUES)
    count = 1

    for D in VECTOR_SIZES:
        for sg in SG_VALUES:

            print(f"\n>>> Experimento {count}/{total}")

            train_and_save(corpus, D, sg)

            count += 1

    print("\nTodos los experimentos finalizados.")

# EJECUCIÓN
if __name__ == "__main__":
    run_experiments(corpus)
```

```python
from gensim.models import KeyedVectors
import pandas as pd
import glob

pd.set_option("display.max_colwidth", 120)
pd.set_option("display.width", 160)

# OUTPUT_DIR
TOPN = 10                     # vecinos a mostrar por prueba
RANDOM_SEED = 42              # reproducibilidad en muestreos
np.random.seed(RANDOM_SEED)
```

## 7 Defienndo las pruebas 

```python

# 2 Analogías médicas (word offset): B - A + C  -> palabra objetivo
ANALOGIES = [
    ("hígado", "hepatitis", "pulmón"),
    ("diabetes", "insulina", "hipertensión"),
]

# 4 términos para similitud (vecinos más cercanos)
SIM_TERMS = ["diabetes", "antibiótico", "hepatitis", "cáncer"]

# 2 conjuntos para intrusos (odd one out)
# Mikolov: vector promedio del grupo y el más distante es el intruso.
ODD_ONE_OUT_SETS = [
    ["ibuprofeno", "paracetamol", "naproxeno", "insulina"],   # analgésicos vs hormona
    ["pulmón", "hígado", "riñón", "amoxicilina"],             # órganos vs antibiótico
]

# 2 términos para "antónimos/alejadas" (menor similitud coseno)
FARTHEST_TERMS = ["benigno", "agudo"]
```

## Funciones

```python
def cosine_sim(u: np.ndarray, v: np.ndarray) -> float:
    """Similitud coseno entre dos vectores."""
    denom = (np.linalg.norm(u) * np.linalg.norm(v))
    if denom == 0:
        return 0.0
    return float(np.dot(u, v) / denom)

def get_vec(kv: KeyedVectors, word: str) -> np.ndarray | None:
    """Devuelve vector o None si OOV."""
    if word not in kv.key_to_index:
        return None
    return kv.get_vector(word)

def most_similar_cosine(kv: KeyedVectors, query_vec: np.ndarray, topn: int = 10, exclude: set[str] | None = None):
    """
    Top-N por coseno usando la implementación optimizada de gensim.
    Si 'exclude' se usa, filtra esas palabras.
    """
    results = kv.similar_by_vector(query_vec, topn=topn + (len(exclude) if exclude else 0))
    if exclude:
        results = [(w, s) for (w, s) in results if w not in exclude]
    return results[:topn]

def word_offset_analogy(kv: KeyedVectors, a: str, b: str, c: str, topn: int = 10):
    """
    Analogía tipo Mikolov: vec(b) - vec(a) + vec(c).
    Devuelve topn predicciones.
    """
    va, vb, vc = get_vec(kv, a), get_vec(kv, b), get_vec(kv, c)
    if va is None or vb is None or vc is None:
        return None, f"OOV en analogía: faltó {', '.join([w for w in [a,b,c] if w not in kv.key_to_index])}"
    query = vb - va + vc
    exclude = {a, b, c}
    preds = most_similar_cosine(kv, query, topn=topn, exclude=exclude)
    return preds, None

def nearest_neighbors(kv: KeyedVectors, term: str, topn: int = 10):
    """Vecinos más cercanos por coseno."""
    v = get_vec(kv, term)
    if v is None:
        return None, f"OOV en similitud: {term}"
    preds = most_similar_cosine(kv, v, topn=topn, exclude={term})
    return preds, None

def odd_one_out_mikolov(kv: KeyedVectors, words: list[str]):
    """
    Detección de intruso estilo Mikolov:
    1) toma vectores del grupo
    2) calcula el promedio (centroide)
    3) el intruso es el de menor similitud coseno con el centroide (más distante)
    """
    oov = [w for w in words if w not in kv.key_to_index]
    if oov:
        return None, f"OOV en intrusos: {', '.join(oov)}"

    vecs = np.vstack([kv.get_vector(w) for w in words])
    centroid = vecs.mean(axis=0)

    sims = [(w, cosine_sim(kv.get_vector(w), centroid)) for w in words]
    sims_sorted = sorted(sims, key=lambda x: x[1])  # menor sim = más distante
    odd_word, odd_sim = sims_sorted[0]
    return {"odd_word": odd_word, "details": sims_sorted}, None

def farthest_words_approx(kv: KeyedVectors, term: str, topn: int = 10, sample_size: int = 50000):
    """
    Palabras alejadas por coseno (aprox):
    - Calcula similitud de 'term' vs un muestreo del vocabulario.
    - Devuelve las topn con menor similitud.
    """
    if term not in kv.key_to_index:
        return None, f"OOV en alejadas: {term}"

    vocab = np.array(list(kv.key_to_index.keys()))
    if len(vocab) <= sample_size:
        sample = vocab
    else:
        idx = np.random.choice(len(vocab), size=sample_size, replace=False)
        sample = vocab[idx]

    vterm = kv.get_vector(term)
    vterm_n = vterm / (np.linalg.norm(vterm) + 1e-12)

    # Tomamos vectores de la muestra
    mats = np.vstack([kv.get_vector(w) for w in sample])
    mats_n = mats / (np.linalg.norm(mats, axis=1, keepdims=True) + 1e-12)

    sims = mats_n @ vterm_n  # coseno
    pairs = list(zip(sample.tolist(), sims.tolist()))
    pairs = [(w, float(s)) for w, s in pairs if w != term]

    pairs_sorted = sorted(pairs, key=lambda x: x[1])  # menor sim primero
    return pairs_sorted[:topn], None

def format_pairs(pairs, k=5):
    """Convierte lista [(w, score), ...] a string"""
    if pairs is None:
        return ""
    return ", ".join([f"{w} ({s:.3f})" for w, s in pairs[:k]])
```

## Carga de los modelos

```python
def parse_model_meta(path: str):
    base = os.path.basename(path)
    m = re.search(r"D(\d+).*sg(\d+)", base)
    D = int(m.group(1)) if m else None
    sg = int(m.group(2)) if m else None
    return os.path.splitext(base)[0], D, sg

model_paths = sorted(glob.glob(os.path.join(OUTPUT_DIR, "*.txt")))
if not model_paths:
    raise FileNotFoundError(f"No encontré modelos .txt en: {OUTPUT_DIR}/")

print("Modelos encontrados:")
for p in model_paths:
    print(" -", os.path.basename(p))

models = {}  # name -> KeyedVectors

t0 = time.time()
for path in model_paths:
    name, D, sg = parse_model_meta(path)
    print(f"\nCargando {name} ...")
    kv = KeyedVectors.load_word2vec_format(path, binary=False)
    models[name] = {"kv": kv, "D": D, "sg": sg, "path": path, "vocab_size": len(kv)}
    print(f"  vocab_size = {len(kv)}")

print(f"\nCarga total: {(time.time()-t0)/60:.2f} min")
```

## Evluacion

```python
def evaluate_model(name: str, kv: KeyedVectors):
    out = {
        "model": name,
        "vocab_size": len(kv),
        "oov_count": 0,
        "analogies_ok": 0,
        "odd_ok": 0,
    }

    details = {
        "analogies": [],
        "similarity": [],
        "odd_one_out": [],
        "farthest": []
    }

    # -------- Analogías (2)
    for (a, b, c) in ANALOGIES:
        preds, err = word_offset_analogy(kv, a, b, c, topn=TOPN)
        if err:
            out["oov_count"] += 1
            details["analogies"].append(f"{a}:{b} :: {c}: ?  ->  {err}")
        else:
            out["analogies_ok"] += 1
            details["analogies"].append(
                f"{a}:{b} :: {c}: ?  ->  {format_pairs(preds, k=5)}"
            )

    # -------- Similitud (4 términos)
    for term in SIM_TERMS:
        preds, err = nearest_neighbors(kv, term, topn=TOPN)
        if err:
            out["oov_count"] += 1
            details["similarity"].append(f"{term} -> {err}")
        else:
            details["similarity"].append(f"{term} -> {format_pairs(preds, k=8)}")

    # -------- Intrusos (2 conjuntos)
    for words in ODD_ONE_OUT_SETS:
        res, err = odd_one_out_mikolov(kv, words)
        if err:
            out["oov_count"] += 1
            details["odd_one_out"].append(f"{words} -> {err}")
        else:
            out["odd_ok"] += 1
            odd = res["odd_word"]
            # mostramos detalles (ordenados por similitud con el centroide)
            det = ", ".join([f"{w}:{s:.3f}" for w, s in res["details"]])
            details["odd_one_out"].append(f"{words} -> intruso: {odd} | sims: {det}")

    # -------- Palabras alejadas / “antónimos” (2)
    # OJO: aproximación con muestreo para no tardar demasiado.
    for term in FARTHEST_TERMS:
        far, err = farthest_words_approx(kv, term, topn=TOPN, sample_size=50000)
        if err:
            out["oov_count"] += 1
            details["farthest"].append(f"{term} -> {err}")
        else:
            details["farthest"].append(f"{term} -> {format_pairs(far, k=8)}")

    # -------- Score 
    # Premia que pueda responder analogías e intrusos y penaliza OOV.
    out["score"] = (out["analogies_ok"] * 3.0) + (out["odd_ok"] * 2.0) - (out["oov_count"] * 1.0)

    return out, details


results = []
reports = {}

t0 = time.time()
for name, meta in models.items():
    kv = meta["kv"]
    print(f"\nEvaluando: {name} (vocab={len(kv)})")
    summary, detail = evaluate_model(name, kv)
    # agrega metadatos si existen
    summary["D"] = meta["D"]
    summary["sg"] = meta["sg"]
    results.append(summary)
    reports[name] = detail

print(f"\nEvaluación total: {(time.time()-t0)/60:.2f} min")

df = pd.DataFrame(results).sort_values(["score", "analogies_ok", "odd_ok", "oov_count"], ascending=[False, False, False, True])
df
```

## Reporte

```python
def print_report(model_name: str):
    print("\n" + "="*90)
    print(f"REPORTE: {model_name}")
    print("="*90)

    det = reports[model_name]

    print("\n[1] Analogías (word offset; Mikolov 2013)")
    for line in det["analogies"]:
        print(" -", line)

    print("\n[2] Similitud (vecinos más cercanos por coseno)")
    for line in det["similarity"]:
        print(" -", line)

    print("\n[3] Detección de intruso (centroide + más distante)")
    for line in det["odd_one_out"]:
        print(" -", line)

    print("\n[4] Palabras alejadas (menor similitud coseno; aproximación por muestreo)")
    for line in det["farthest"]:
        print(" -", line)

# Imprime reportes en el orden del ranking
for model_name in df["model"].tolist():
    print_report(model_name)
```
